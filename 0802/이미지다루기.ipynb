## 이미지에 대한 색상 구분하기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr = cv2.imread('c:/pythonworkspace/multi/0802/data/plane.jpg')

# BGR에서 HSV로 변환 참고 : https://www.charlezz.com/?p=44906
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
# img_hsv

# HSV에서 파란색 값 범위 정의
lower_blue = np.array([50, 100, 50])
upper_blue = np.array([130, 255, 255])


# 마스크 생성
mark = cv2.inRange(img_hsv, lower_blue, upper_blue)

# 마스크 적용
img_hsv_mark = cv2.bitwise_and(img_bgr, img_hsv, mark)
# img_hsv_mark

img_rgb = cv2.cvtColor(img_hsv_mark, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb, cmap = 'gray')
plt.show()

plt.imshow(img_hsv_mark, cmap = 'gray')
plt.show()
```

## 배경제거

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_bgr = cv2.imread('c:/pythonworkspace/multi/0802/data/plane.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 사각형 좌표 : 시작점 x, y 너비/높이 w, h
rectag = (700, 500, 2650, 1200)

# 마스크 값 생성
mask = np.zeros(img_rgb.shape[:2], np.uint8)
```

```python
bgrmod = np.zeros((1, 65), np.float64)
brgmod = np.zeros((1, 65), np.float64)
```

```python
cv2.grabCut(img_rgb,                # 원본데이터
            mask,                   # 마스크
            rectag,                 # 사각형
            bgrmod,                 # 배경을 위한 임시 배열
            brgmod,                 # 전경을 위한 임시 배열
            5,                      # 반복횟수
            cv2.GC_INIT_WITH_RECT)  # 

mask_2 = np.where((mask==2)|(mask==0), 0, 1).astype('uint8')
img_nobg = img_rgb*mask_2[:, :, np.newaxis]

plt.imshow(img_nobg)
plt.show()
```

## 경계선 감지

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread('c:/pythonworkspace/multi/0802/data/plane_1.jpg', cv2.IMREAD_GRAYSCALE)

# 픽셀강도의 중간값을 계산
median_intensity = np.median(img_gray)
print(median_intensity)

# 중간 픽셀 강도에서 위아래 1 표준편차 떨어진 값을 임곗값으로 지정

lower_threshold = int(max(0, (1.0 - 0.33) * median_intensity))
upper_threshold = int(min(255, (1.0 + 0.33) * median_intensity))

# Canny() 메소드: 그레디언트 임곗값 사이의 저점과 고점을 나타내는 두 매개변수가 필요
# 낮은 임곗값과 높은 임곗값 사이의 가능성 있는 경계선 픽셀은 약한 경계선 픽셀로 간주
# 높은 임곗값보다 큰 픽셀은 강한 경계선 픽셀로 간주
img_canny = cv2.Canny(img_gray, lower_threshold, upper_threshold)

plt.imshow(img_canny, cmap = 'gray')
plt.show()

plt.imshow(img_gray, cmap='gray')
plt.show()
```
